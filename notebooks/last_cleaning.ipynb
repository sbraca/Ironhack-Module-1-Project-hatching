```python
import requests
!pip install BeautifulSoup4
!pip install lxml
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#setting URL and getting all content
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_tertiary_education_attainment'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
```

```python
#finding tertiary education list
table = soup.find_all('table',{'class':'wikitable sortable'})[0]
table
```

```python
#parsing rows
rows = table.find_all('tr')
rows_parsed = [row.text for row in rows]
rows_parsed
```

```python
#nested list/ rows
rows = [row.text.strip().split("\n") for row in rows]
rows
```

```python
#specify columns and rows to form a DataFrame
colnames = rows[0]
print(colnames)

for ele in colnames: 
    if ele == (''): 
        colnames.remove(ele) 

print(colnames)
```

```python
colnames = ['Country', 'Age25–64(%)', 'Age25-34(%)','Age35-44(%)', 'Age45-54(%)', 'Age55-64(%)', 'Year']
```

```python
colnames
```

```python
data = rows[1:]

tertiary_ed = pd.DataFrame(data, columns=colnames)
tertiary_ed
```

```python
tertiary_ed.drop([0], inplace=True) 
```

```python
tertiary_ed.drop(['Age25-34(%)','Age35-44(%)','Age45-54(%)', 'Age55-64(%)', 'Year'],  axis=1, inplace=True)
```
